## Reconstruction
Evaluate the ability of this algorithm to reconstruct the input motion (evaluation method used in aberman et al. (2020))

```python
# tested using the lafan dataset
# get same input and adversarial dataset
input_dataset = UnityMotionDataset(motion_path_input, options=options)
adv_dataset = UnityMotionDataset(motion_path_output, options=options)
skdata_input = input_dataset.skdata
skdata_adv = adv_dataset.skdata

# test_input_inds = np.random.randint(0,len(self.input_dataset) - self.options["window_size"], test_size)
start_indices = np.array([100,116,132])
motion = input_dataset.to_skaware(start_indices, start_indices + options["window_size"])
real_motion = adv_dataset.to_skaware(start_indices, start_indices + options["window_size"])


motion_copy = motion.clone()
# add normal noise
# motion = motion + (torch.rand_like(motion)*0.01)

# add dropout: randomly zero out some joint values?
dropout = torch.nn.Dropout(p=0.1)
motion = dropout(motion)
# normalize (not necessary)
motion = motion.reshape(3,16,skdata_adv.num_joints, 7)
rotations = motion[:,:,:4].clone()
rotations = torch.nn.functional.normalize(rotations, dim=3)
motion[:,:,:4] = rotations
motion = motion.reshape(3, 16, -1)

# get the retargetted motion
fake_motion_data, real_motion_data, input_motion_data = trainer.retarget(motion, real_motion)

# extract motion and velocity
input_motion_data = utils.get_pos_info_from_raw(motion_copy, trainer.skdata_input, options)
input_motion_cat = input_motion_data[0].reshape(-1,skdata_adv.num_joints,3)
fake_motion_cat = fake_motion_data[1].reshape(-1,skdata_adv.num_joints,3)
adv_motion_cat = real_motion_data[1].reshape(-1,skdata_adv.num_joints,3)

input_velocity_cat = input_motion_data[4].reshape(-1,skdata_adv.num_joints,3)
fake_velocity_cat = fake_motion_data[4].reshape(-1,skdata_adv.num_joints,3)
adv_velocity_cat = real_motion_data[4].reshape(-1,skdata_adv.num_joints,3)

# apply rotation offset for the velocity
curr_batch_size = input_motion_cat.shape[0]
rot_offset = torch.tensor([0.707,0.,0.707,0.]).float()
vel_rot_offset = rot_offset.reshape(1,1,4).repeat(curr_batch_size, skdata_input.num_joints, 1)

input_velocity_cat = utils.quat_mul_vec(vel_rot_offset, input_velocity_cat)
fake_velocity_cat = utils.quat_mul_vec(vel_rot_offset, fake_velocity_cat)

# retrieve the global position by intergrating the root velocity
# _ , input_glob_pos = utils.get_global_position_from_velocity(torch.zeros(3), input_velocity_cat[:,0], skdata_input.frametime, positions= input_motion_cat)
# _ , fake_glob_pos = utils.get_global_position_from_velocity(torch.zeros(3), fake_velocity_cat[:,0], skdata_adv.frametime, positions= fake_motion_cat)
```

```python
# calculate loss in the same way as with the paper
height = skdata_input.height
print(torch.mean((input_motion_cat/height - fake_motion_cat/height )**2)*1000)
```

```python
# display the resutls
curr_batch_size = input_motion_cat.shape[0]
rot_offset = torch.tensor([0.707,0.707,0.,0.]).float()
pos_rot_offset = rot_offset.reshape(1,1,4).repeat(curr_batch_size, skdata_input.num_joints, 1)

input_motion_cat = utils.quat_mul_vec(pos_rot_offset, input_motion_cat) # local position
fake_motion_cat = utils.quat_mul_vec(pos_rot_offset, fake_motion_cat) # local position

input_glob_pos = utils.quat_mul_vec(pos_rot_offset, input_glob_pos) # global position
fake_glob_pos = utils.quat_mul_vec(pos_rot_offset, fake_glob_pos) # global position

limits=[[-1,1],[-1,1],[-1,1]]
# anim = motion_animation([input_motion_cat.detach().numpy(), fake_motion_cat.detach().numpy()], [skdata_adv.edges,skdata_adv.edges], ['g', 'r'], limits=limits, framerate=200)
anim = motion_animation([input_motion_cat.detach().numpy(), fake_motion_cat.detach().numpy()], [skdata_adv.edges, skdata_adv.edges], ['g', 'r'], limits=limits, framerate=200)
# anim.save("C:/users/nicol/Work/Master/dissertation/results/LaFan_to_LaFan_dropout_with_vel.gif", writer=PillowWriter(fps=15))
# anim.save("C:/users/nicol/Work/Master/dissertation/results/LaFan_to_LaFan_no_vel.gif", writer=PillowWriter(fps=15))
# HTML(anim.to_jshtml())
```

## Retargeting Evaluation
Retargeting from one character to another. 
Retargeting from Ortiz to James, James to Ortiz -> ground truth available for quantitative results
Retargeting from Ortiz to LaFan -> different datasets, ground truth not available.

```python
motion_path_input = "./data/Ortiz/"
motion_path_output = "./data/LaFan/"

# tested using the lafan dataset
# get same input and adversarial dataset
input_dataset = UnityMotionDataset(motion_path_input, options=options)
adv_dataset = UnityMotionDataset(motion_path_output, options=options)
skdata_input = input_dataset.skdata
skdata_adv = adv_dataset.skdata

# test_input_inds = np.random.randint(0,len(self.input_dataset) - self.options["window_size"], test_size)
# start_indices = np.array([66, 82, 98])
start_indices = np.array([100,108,116,124])

motion = input_dataset.to_skaware(start_indices, start_indices + options["window_size"])
real_motion = adv_dataset.to_skaware(start_indices, start_indices + options["window_size"])

# get the retargetted motion
fake_motion_data, real_motion_data, input_motion_data = trainer.retarget(motion, real_motion)


# extract motion and velocity
input_motion_cat = input_motion_data[0][:,8:].reshape(-1,skdata_input.num_joints,3)
fake_motion_cat = fake_motion_data[1][:,8:].reshape(-1,skdata_adv.num_joints,3)
adv_motion_cat = real_motion_data[1][:,8:].reshape(-1,skdata_adv.num_joints,3)

input_velocity_cat = input_motion_data[4].reshape(-1,skdata_input.num_joints,3)
fake_velocity_cat = fake_motion_data[4].reshape(-1,skdata_adv.num_joints,3)
adv_velocity_cat = real_motion_data[4].reshape(-1,skdata_adv.num_joints,3)

input_motion_cat[:,:,0] -= 0.3
fake_motion_cat[:,:,0] += 0.3
adv_motion_cat[:,:,0] += 0.3
```
